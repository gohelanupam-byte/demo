```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('homeprices.csv')
```

```python
df
```

```python
plt.scatter(df.area,df.price)
plt.xlabel('area')
plt.ylabel('price')
```

```python
import sklearn.linear_model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
line = LinearRegression()
```

```python
line.fit(df[['area']], df['price'])
```

```python
line.coef_
```

```python
line.intercept_
```

```python
line.predict([[5000]])
```

```python
df1 = pd.read_csv('canada_per_capita_income.csv')
```

```python
df1
```

```python
line1 = LinearRegression()
```

```python
line1.fit(df1[['year']], df1['per capita income (US$)'])
```

```python
plt.scatter(df1.year,df1['per capita income (US$)'])
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.plot(df1['year'], line1.predict(df1[['year']]), color='red')
```

```python
line1.coef_
```

```python
line1.intercept_
```

```python
line1.predict([[2025]])
```

```python
plt.plot(df1['year'], line1.predict(df1[['year']]))
```
